```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.decomposition import PCA

%matplotlib inline
```

```python
X_train_all = np.load(open('X_train.npy', 'rb'))
Y_train_all = np.load(open('Y_train.npy', 'rb'))
X_test_all = np.load(open('X_test.npy', 'rb'))
Y_test_all = np.load(open('Y_test.npy', 'rb'))

#organize the data by category/class
X_train1 = X_train_all[:200, :]
X_train2 = X_train_all[200:400, :]
X_train3 = X_train_all[400:600, :]
X_train4 = X_train_all[600:800, :]
X_train5 = X_train_all[800:1000, :]
X_train = np.concatenate((X_train1, X_train2, X_train3, X_train4, X_train5), axis=0)
X_train = X_train.reshape(X_train.shape[0], -1)

Y_train1 = Y_train_all[:200, :]
Y_train2 = Y_train_all[200:400, :]
Y_train3 = Y_train_all[400:600, :]
Y_train4 = Y_train_all[600:800, :]
Y_train5 = Y_train_all[800:1000, :]
Y_train = np.concatenate((Y_train1, Y_train2, Y_train3, Y_train4, Y_train5), axis=0)
Y_train = np.argmax(Y_train, axis=1)

X_test1 = X_test_all[:40, :]
X_test2 = X_test_all[40:80, :]
X_test3 = X_test_all[80:120, :]
X_test4 = X_test_all[120:160, :]
X_test5 = X_test_all[160:200, :]
X_test = np.concatenate((X_test1, X_test2, X_test3, X_test4, X_test5), axis=0)
X_test = X_test.reshape(X_test.shape[0], -1)

Y_test1 = Y_test_all[:40, :]
Y_test2 = Y_test_all[40:80, :]
Y_test3 = Y_test_all[80:120, :]
Y_test4 = Y_test_all[120:160, :]
Y_test5 = Y_test_all[160:200, :]
Y_test = np.concatenate((Y_test1, Y_test2, Y_test3, Y_test4, Y_test5), axis=0)
Y_test = np.argmax(Y_test, axis=1)

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print('X_test shape:', X_test.shape)
print('Y_test shape:', Y_test.shape)
```

```python
#model call with processed data
clf = svm.SVC(kernel='poly', gamma='scale')
clf.fit(X_train, Y_train)

trainAcc = clf.score(X_train, Y_train)

print('SVM training accuracy:', trainAcc)

testAcc = clf.score(X_test, Y_test)

print('SVM testing accuracy:', testAcc)
```

```python
# Plot data and support vectors
# PCA projection
pca = PCA(n_components=2)
pca.fit(X_train)
X_proj = pca.transform(X_train)

# Start plotting
fig = plt.figure(figsize=(8,8))

# Plot data points
plt.scatter(X_proj[:200, 0], 
            X_proj[:200, 1], c='dodgerblue', label="Drawings")
plt.scatter(X_proj[200:400, 0], 
            X_proj[200:400, 1], c='salmon', label="Engravings")
plt.scatter(X_proj[400:600, 0], 
            X_proj[400:600, 1], c='springgreen', label="Iconography")
plt.scatter(X_proj[600:800, 0], 
            X_proj[600:800, 1], c='gold', label="Paintings")
plt.scatter(X_proj[800:1000, 0], 
            X_proj[800:1000, 1], c='lightgray', label="Sculptures")

# Plot support vectors (the darker colors will overlap those points that are already plotted)
plt.scatter(X_proj[clf.support_[0:clf.n_support_[0]],0], X_proj[clf.support_[0:clf.n_support_[0]],1], c='darkblue')
plt.scatter(X_proj[clf.support_[clf.n_support_[0]:np.sum(clf.n_support_[:2])],0], 
            X_proj[clf.support_[clf.n_support_[0]:np.sum(clf.n_support_[:2])],1], c='darkred')
plt.scatter(X_proj[clf.support_[np.sum(clf.n_support_[:2]):np.sum(clf.n_support_[:3])],0], 
            X_proj[clf.support_[np.sum(clf.n_support_[:2]):np.sum(clf.n_support_[:3])],1], c='darkgreen')
plt.scatter(X_proj[clf.support_[np.sum(clf.n_support_[:3]):np.sum(clf.n_support_[:4])],0], 
            X_proj[clf.support_[np.sum(clf.n_support_[:3]):np.sum(clf.n_support_[:4])],1], c='darkgoldenrod')
plt.scatter(X_proj[clf.support_[np.sum(clf.n_support_[:4]):np.sum(clf.n_support_[:5])],0], 
            X_proj[clf.support_[np.sum(clf.n_support_[:4]):np.sum(clf.n_support_[:5])],1], c='dimgray')
plt.legend()
plt.title("Feature Projections and Support Vectors")
plt.xlabel("Training Features Projected on X")
plt.ylabel("Training Features Projected on Y")
plt.show()
```
